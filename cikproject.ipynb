```python
import numpy as np
import pandas as pd
```

```python
links=pd.read_csv("web_url.csv",dtype=str)
links['cik']=links['cik']+"-"+links['date']
cik=links['cik'].tolist()
link=links['link'].tolist()
```

```python
dict={}
for id,url in zip(cik,link):
    dict[id]=url
dict
```

```python
Tables={}
for cik in dict:
	print (cik)
	url=dict[cik]
	try:
		Tables[cik]=pd.read_html(url,encoding = "ISO-8859-1")
	except (ValueError,HTTPError):
		continue
```

```python
url=dict['1084384-2013-11-21']
Tables['1084384-2013-11-21']=pd.read_html(url,encoding = "ISO-8859-1")
```

```python
import urllib.request
import os
from urllib.error import HTTPError
os.mkdir(".\\Summary Compensation Table")  
for item in dict:
    try:
        urllib.request.urlretrieve(dict[item],".\\Summary Compensation Table\\"+str(item)+'.htm')
    except HTTPError:
        print (item)
        continue      
```

```python
len(Tables.keys())
```

```python
final={} 
keyword=["name","principal","position"]
for cik in Tables:
	test=Tables[cik]
	for table in test:
		checklist=[]
		table=table.dropna(axis=0,how="all").dropna(axis=1,how="all")
		try:
			column=np.asarray(table.iloc[:,0])
		except IndexError:
			continue
		for item in column:
			try:
				item=item.replace("\xa0"," ")
			except AttributeError:
				pass
			checklist.append(item)
		list_str=str(checklist).lower()
		if keyword[0] in list_str and (keyword[1] in list_str or "principle" in list_str) and keyword[2] in list_str:
			table.to_csv("./Summary Compensation Table/"+str(cik)+".csv",index=False)
			final[cik]=table
```
