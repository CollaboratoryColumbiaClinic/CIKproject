```python
import numpy as np
import pandas as pd
```

```python
links=pd.read_csv("fraction.csv",header=None)
links=links[0].tolist()
```

```python
dict={}
for url in links:
    url=url.split("\t")
    link="https://www.sec.gov"+url[0][3:-2]
    id=url[1]
    dict[id]=link    
dict
```

```python
Tables={}
for cik in dict:
	url=dict[cik]
	Tables[cik]=pd.read_html(url)
```

```python
import urllib.request
for item in dict:
    urllib.request.urlretrieve(dict[item],"./Summary Compensation Table/"+item+".htm")
    
```

```python
test=Tables['43300']
num=0
table=test[80]
table=table.dropna(axis=0,how="all").dropna(axis=1,how="all")
table.to_csv("table.csv")
```

```python
final={} 
keyword=["Name","Principal","Position"]
for cik in Tables:
	test=Tables[cik]
	for table in test:
		list=[]
		table=table.dropna(axis=0,how="all").dropna(axis=1,how="all")
		try:
			column=np.asarray(table.iloc[:,0])
		except IndexError:
			continue
		for item in column:
			try:
				item=item.replace("\xa0"," ")
			except AttributeError:
				continue
			list.append(item)
		if keyword[0] in str(list) and keyword[1] in str(list) and keyword[2] in str(list):
			table.to_csv("./Summary Compensation Table/"+cik+".csv",index=False)
			final[cik]=table
```
